**a)Create and draw a 6-qubit/6-classical bit quantum circuit where you apply an X followed by H gate to each qubit using a for loop and measure all the bits.**

```python
# Define quantum circuit
qc = QuantumCircuit(6, 6)  # Creating a quantum circuit with 6 qubits and 6 classical bits

# Apply X followed by H gate to each qubit using a for loop
for qubit in range(6):
    qc.x(qubit)  # Apply X gate on the current qubit
    qc.h(qubit)  # Apply H gate on the current qubit

# Measure all qubits
qc.measure(range(6), range(6))  # Measure qubits 0 to 5 and store the results in classical bits 0 to 5

# Draw
qc.draw('mpl')
```

**b) Create and draw a 5-qubit/5-classical bit quantum circuit where you apply the gates listed below using a loop and measure.**

H gate on qubit 0

CX with qubit 0 as the control and qubit 1 as the target

H gate on qubit 1

CX with qubit 1 as the control and qubit 2 as the target

H gate on qubit 2

CX with qubit 2 as the control and qubit 3 as the target

```python
# Define quantum circuit
qc = QuantumCircuit(5, 5)  # Creating a quantum circuit with 5 qubits and 5 classical bits

# Apply H gate and CX gate using a loop
for qubit in range(4):
    qc.h(qubit)         # Apply H gate on the current qubit
    qc.cx(qubit, qubit+1)  # Apply CX gate with the current qubit as control and the next qubit as target

# Apply H gate on the last qubit
qc.h(4)

# Measure all qubits
qc.measure(range(5), range(5))  # Measure qubits 0 to 4 and store the results in classical bits 0 to 4

# Draw
qc.draw('mpl')
```
